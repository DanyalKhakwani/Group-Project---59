## API Setup

```python
headers = {
    "Authorization": f"Bearer {github_token}",
    "Accept": "application/vnd.github+json"
}
 
def get(url, params=None):
    response = requests.get(url, headers=headers, params=params)
    print(f" Requests remaining: {response.headers.get('X-RateLimit-Remaining')}")
    return response.json()
 
base_pandas = f"https://api.github.com/repos/pandas-dev/pandas"
base_modin = f"https://api.github.com/repos/modin-project/modin"
```

## PR Meta Data
Usecase 1: This meta data will be used for basic analysis of community health. Mostly EDA on number of prs/day etc.

Usecase 2: The 'body' column of PRs will be used for Topic Modelling

```python
def pr_collection(base):
    """
    This function takes in a URL, for pandas or modin, and then pulls metadata for 1k repos, returns a dataframe.
    """
    prs = []
    pages = 1
     
    while len(prs) < total_prs:
        batch = get(f"{base}/pulls", params={
            "state":    "closed",
            "per_page": 100,
            "page":     pages
        })
        if not batch:
            break
        prs.extend(batch)
        print(f" Page number: {pages}, PR fetched: {len(batch)}, total: {len(prs)}")
        pages += 1
     
    prs = prs[:total_prs]
     
    df_prs = pd.DataFrame([{
        "pr_id":  pr["number"],
        "pr_title":      pr["title"],
        "pr_author":     pr["user"]["login"],
        "state":      pr["state"],
        "created_at": pr["created_at"],
        "merged_at":  pr.get("merged_at"),
        "body":       pr.get("body", "")
    } for pr in prs])

    return df_prs
```

### Pandas

```python
df_prs_pandas = pr_collection(base_pandas)
```

## Reviewer Data
Usecase: This shows the relationship in the community between author of the pull request and reviewer, to check the contribution of the overall community

```python
def collect_reviews(df, base):
    """
    This function collects pr author and reviewer relationship for the most recent 300 prs. Returns a dataframe.
    """
    review = []
    prs = df.head(300)
     
    for i, row in enumerate(prs.itertuples()):
        reviews = get(f"{base}/pulls/{row.pr_id}/reviews")

        for r in reviews:
            if not r.get("user"):
                continue
            review.append({
                "pr_id":        row.pr_id,
                "pr_author":    row.pr_author,
                "reviewed_by":  r["user"]["login"],
                "review_state": r["state"],
                "review_text":  r["body"],
                "reviewed_at":  r["submitted_at"],
            })
     
        if i % 25 == 0:
            print(f"  Progress: {i}/{len(prs)} PRs")
        time.sleep(0.5)
     
    df_reviews = pd.DataFrame(review)

    return df_reviews
```

```python
df_reviews_pandas = collect_reviews(df_prs_pandas, base_pandas)
```
